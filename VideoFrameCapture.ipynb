```python
PATH = r"C:\Users\WIN 8.1\Downloads\22.mp4"
videocapture = cv2.VideoCapture(PATH)
success, image = videocapture.read()
count = 0
success = True
while(success):
    cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
    success,image = videocapture.read()
    print('Read a new frame: ', success)
    count += 1
```
